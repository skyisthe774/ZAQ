```python
# 고유 클래스 이름을 숫자로 매핑
class_to_idx = {label: idx for idx, label in enumerate(origin_train_df['label'].unique())}
origin_train_df['label'] = origin_train_df['label'].map(class_to_idx)

print(origin_train_df['label'].unique())  # 숫자로 변환되었는지 확인
```

```python
class CustomImageDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.dataframe = dataframe
        self.transform = transform

    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        img_path = self.dataframe.iloc[idx]['path']
        label = self.dataframe.iloc[idx]['label']  # 이제 label은 숫자형
        image = Image.open(img_path).convert("RGB")
        if self.transform:
            image = self.transform(image)
        return image, torch.tensor(label, dtype=torch.long)  # 텐서로 변환
```

```python
train_df, val_df = train_test_split(origin_train_df, test_size=0.2, random_state=42, stratify=origin_train_df['label'])

train_dataset = CustomImageDataset(train_df, transform=transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # ImageNet 기준 정규화
]))

val_dataset = CustomImageDataset(val_df, transform=transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
]))

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False, num_workers=4)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Pretrained ResNet 모델 로드
model = models.resnet18(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(origin_train_df['label'].unique()))  # 클래스 수에 맞게 설정
model = model.to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
for images, labels in tqdm(train_loader):
    images = images.to(device)  # 이미지 텐서
    labels = labels.to(device)  # 레이블 텐서

    optimizer.zero_grad()
    outputs = model(images)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
```

```python
num_epochs = 10

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in tqdm(train_loader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    train_accuracy = 100 * correct / total
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader)}, Accuracy: {train_accuracy:.2f}%")

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    val_accuracy = 100 * val_correct / val_total
    print(f"Validation Accuracy: {val_accuracy:.2f}%")
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Validation 데이터 전체를 사용하여 예측
all_preds = []
all_labels = []
model.eval()

with torch.no_grad():
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Classification Report
print(classification_report(all_labels, all_preds))
```

```python
# 잘못된 예측 인덱스 확인
misclassified_indices = [i for i, (true, pred) in enumerate(zip(all_labels, all_preds)) if true != pred]

# 첫 번째 잘못된 예측 샘플 출력 (예시)
print(f"True Label: {all_labels[misclassified_indices[0]]}, Predicted: {all_preds[misclassified_indices[0]]}")
misclassified_image = val_dataset[misclassified_indices[0]][0].permute(1, 2, 0)  # 이미지 형식 맞추기
plt.imshow(misclassified_image)
plt.show()
```

```python
train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
```

```python
from torchvision.models import resnet50

model = resnet50(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, len(origin_train_df['label'].unique()))  # 클래스 수 맞추기
model = model.to(device)
```

```python
# 손실 함수와 옵티마이저 설정
criterion = nn.CrossEntropyLoss()  # 분류 문제에 적합한 손실 함수
optimizer = optim.Adam(model.parameters(), lr=0.001)  # Adam 옵티마이저 사용
```

```python
import matplotlib.pyplot as plt

# 손실 값을 기록할 리스트들 (train_losses, valid_losses)
train_losses = []
valid_losses = []

num_epochs = 10  # 학습할 에포크 수

for epoch in range(num_epochs):
    model.train()  # 모델을 학습 모드로 전환
    running_train_loss = 0.0
    correct = 0
    total = 0

    # 학습 데이터셋에 대해 학습
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)  # 데이터 GPU로 전송

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass 및 옵티마이저 업데이트
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # 손실 및 정확도 계산
        running_train_loss += loss.item()
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # 학습 손실 기록
    train_losses.append(running_train_loss / len(train_loader))

    # 학습 정확도 출력
    train_accuracy = 100 * correct / total
    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {running_train_loss/len(train_loader):.4f}, Train Accuracy: {train_accuracy:.2f}%")

    # 검증 데이터셋에 대해 검증
    model.eval()  # 모델을 평가 모드로 전환
    running_valid_loss = 0.0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)

            running_valid_loss += loss.item()
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()

    # 검증 손실 기록
    valid_losses.append(running_valid_loss / len(val_loader))

    # 검증 정확도 출력
    val_accuracy = 100 * val_correct / val_total
    print(f"Validation Loss: {running_valid_loss/len(val_loader):.4f}, Validation Accuracy: {val_accuracy:.2f}%")

# 실제 에포크 수 (학습 종료 후에 기록된 손실의 길이)
actual_epochs = len(train_losses)

# 하나의 그래프에 Train Loss와 Validation Loss를 겹쳐서 표시
plt.figure(figsize=(8, 6))

# Train Loss 그래프
plt.plot(range(1, actual_epochs + 1), train_losses, label='Train Loss', color='blue', marker='o')

# Validation Loss 그래프
plt.plot(range(1, actual_epochs + 1), valid_losses, label='Validation Loss', color='orange', marker='x')

# 그래프 제목과 레이블 설정
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Train and Validation Loss')

# 범례 추가
plt.legend()

# 그래프 출력
plt.tight_layout()
plt.show()
```
